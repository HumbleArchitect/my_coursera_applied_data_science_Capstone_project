```python
## Task 1: Mark all launch sites on a map

## Task 1: Mark all launch sites on a map

# First, let's make sure we have all the necessary data and imports
import folium
import pandas as pd
from folium.plugins import MarkerCluster
from folium.plugins import MousePosition
from folium.features import DivIcon

# Download and read the data
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Select relevant sub-columns
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]

print("Launch Sites DataFrame:")
print(launch_sites_df)

# Define NASA Johnson Space Center coordinates
nasa_coordinate = [29.559684888503615, -95.0830971930759]

# Initial the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# For each launch site, add a Circle object based on its coordinate (Lat, Long) values. 
# In addition, add Launch site name as a popup label
for index, site in launch_sites_df.iterrows():
    site_coordinate = [site['Lat'], site['Long']]
    
    # Create a circle with popup label
    circle = folium.Circle(
        site_coordinate, 
        radius=1000, 
        color='#000000', 
        fill=True
    ).add_child(folium.Popup(site['Launch Site']))
    
    # Create a marker with text label
    marker = folium.map.Marker(
        site_coordinate,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % site['Launch Site'],
        )
    )
    
    site_map.add_child(circle)
    site_map.add_child(marker)

# Display the map
site_map
```

```python
## TASK 3: Calculate the distances between a launch site to its proximities

# First, define the calculate_distance function
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Now let's work with CCAFS LC-40 launch site as an example
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Example coordinates for different proximities (you can get these by hovering over the map)
coastline_lat, coastline_lon = 28.56367, -80.56815  # Closest coastline point
city_lat, city_lon = 28.612, -80.807  # Example city (Titusville)
highway_lat, highway_lon = 28.563, -80.655  # Example highway point
railway_lat, railway_lon = 28.572, -80.585  # Example railway point

# Calculate distances
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)

print(f"Distance to coastline: {distance_coastline:.2f} km")
print(f"Distance to city: {distance_city:.2f} km")
print(f"Distance to highway: {distance_highway:.2f} km")
print(f"Distance to railway: {distance_railway:.2f} km")

# TODO: Draw a PolyLine between a launch site to the selected coastline point
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate
coordinates = [[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]]
lines = folium.PolyLine(locations=coordinates, weight=2, color='blue')
site_map.add_child(lines)

# TODO: Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site

# Add distance marker for coastline
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % f"{distance_coastline:.2f} km to Coastline",
    )
)
site_map.add_child(distance_marker)

# Similarly, add lines and markers for other proximities

# City
coordinates_city = [[launch_site_lat, launch_site_lon], [city_lat, city_lon]]
lines_city = folium.PolyLine(locations=coordinates_city, weight=2, color='green')
site_map.add_child(lines_city)

distance_marker_city = folium.Marker(
    [city_lat, city_lon],
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % f"{distance_city:.2f} km to City",
    )
)
site_map.add_child(distance_marker_city)

# Highway
coordinates_highway = [[launch_site_lat, launch_site_lon], [highway_lat, highway_lon]]
lines_highway = folium.PolyLine(locations=coordinates_highway, weight=2, color='red')
site_map.add_child(lines_highway)

distance_marker_highway = folium.Marker(
    [highway_lat, highway_lon],
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % f"{distance_highway:.2f} km to Highway",
    )
)
site_map.add_child(distance_marker_highway)

# Railway
coordinates_railway = [[launch_site_lat, launch_site_lon], [railway_lat, railway_lon]]
lines_railway = folium.PolyLine(locations=coordinates_railway, weight=2, color='orange')
site_map.add_child(lines_railway)

distance_marker_railway = folium.Marker(
    [railway_lat, railway_lon],
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % f"{distance_railway:.2f} km to Railway",
    )
)
site_map.add_child(distance_marker_railway)

# Display the updated map
site_map
```
